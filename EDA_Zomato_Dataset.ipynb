```python
# Relationship between the attributes

rating=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'COUNT'})
```

```python
final_df.columns
```

```python
rating.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)

sns.barplot(x="Aggregate rating",y="COUNT",data=rating)
```

Observation:

Not Rated count is very high
Maximum number of rating are between 2.6 to 3.8

```python

## Count plot
sns.countplot(x="Rating color",data=rating,palette=['blue','red','orange','yellow','green','green'])
```
