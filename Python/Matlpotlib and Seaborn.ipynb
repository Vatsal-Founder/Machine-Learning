## MatplotLib Tutorial
Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

Generally easy to get started for simple plots
Support for custom labels and texts
Great control of every element in a figure
High-quality output in many formats
Very customizable in general

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline #this is use to display the graph other way is my using plt.show()
```

```python
# Scatter Plot

x=np.arange(0,10)
y=np.arange(11,21)

plt.scatter(x,y,s=12,c='r')  ##s is size and c is color
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

```python
# Normal Plot
```

```python
y=x*x
```

```python
plt.plot(x,y,'r') ##r is red color
```

```python
plt.plot(x,y,'g*',linestyle='dashed',linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

```python
## Creating Subplots - this will create subplot which means (2,2) means 2 rows and 2 colums (2,2,1) 1 is the position of the graph

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
#Subplot()
# Compute the x and y coordinates for points on sine and cosine curves 
x = np.arange(0, 5 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2, 1, 1)
   
# Make the first plot 
plt.plot(x, y_sin,'r--') 
plt.title('Sine')  
   
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'g--') 
plt.title('Cosine')  
   
# Show the figure. 
plt.show()
```

```python
## Bar plot

x = [2,8,10] 
y = [11,16,9]  

x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x,y,color='r') 
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  

plt.show()
```

## Seaborn Tutorial
Distribution plots
1. distplot
2. joinplot
3. pairplot


```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
sns.heatmap(df.corr())
```

## Join Plot
A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot


```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

## Pair Plot
A “pairs plot” is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables. This is used for more than 2 features

```python
sns.pairplot(df) ##it do correlation between all the interger values
```

```python
sns.pairplot(df,hue='sex') ##hue allows to differentiate / classify the plot 
```

## Dist plot
Dist plot helps us to check the distribution of the columns feature - this is used for single feature- univariate analysis

```python
sns.distplot(df['tip']) #histogram shows the density of x(element) in y axis
```

```python
sns.distplot(df['tip'],kde=False,bins=10) #kde =density distribution---the current graph gives the count of (x axis)
```
